1.
Write a user-defined function that takes a number as input and returns:
•
"Positive Even"
•
"Positive Odd"
•
"Negative"
•
"Zero"
Use proper conditional statements and explain your logic.

```python
def classify_number(num):
    if num > 0:
        if num % 2 == 0:
            return "Positive Even"
        else:
            return "Positive Odd"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"
```

```python
print(classify_number(10))
print(classify_number(7))
print(classify_number(-5))
print(classify_number(0))
```

2.Write a function that accepts three numbers and returns the largest number without using built-in functions like max(). Explain how your conditional structure ensures correctness.

```python
def largest_number(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
```

```python
print(largest_number(10, 20, 30))
```

3.Write a function that accepts a student’s marks and returns the grade based on the following logic: • 90+ → A • 75–89 → B • 50–74 → C • Below 50 → Fail Use nested conditional statements and justify your approach.

```python
def assign_grade(marks):
    if marks >= 90:
        return "A"
    else:
        if marks >= 75:
            return "B"
        else:
            if marks >= 50:
                return "C"
            else:
                return "Fail"
```

```python
print(assign_grade(85))
print(assign_grade(70))
print(assign_grade(50))
print(assign_grade(30))
```

4.Write a function that takes age as input. • If age is negative, raise an error. • If age is below 18, return "Minor", otherwise return "Adult". Explain how conditional statements help in input validation.

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    elif age < 18:
        return "Minor"
    else:
        return "Adult"
```

```python
print(check_age(15))
print(check_age(20))
```

5.Write a function that accepts a list of integers and: • Counts even numbers • Ignores negative numbers • Stops processing if zero appears Use both loops and conditional statements inside the function.

```python
def process_numbers(numbers):
    even_count = 0

    for num in numbers:
        if num == 0:
            break          # Stop processing if zero appears
        
        if num < 0:
            continue       # Ignore negative numbers
        
        if num % 2 == 0:
            even_count += 1

    return even_count
```

```python
print(process_numbers([1, 2, 3,4, 0, 5, 6]))
```
