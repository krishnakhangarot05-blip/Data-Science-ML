4. Using a single list comprehension, generate a list of squares of numbers from 1 to 20 that are divisible by both 2 and 5.

```python
items = [10, -3, "abc", 5, 0, 7]

for item in items:
    try:
        item = int(item)

        if item < 0:
            continue
        if item == 0:
            print("Stopped at zero")
            break

        print(item)

    except ValueError:
        print("Non-integer value encountered")
```

```python
import sys

num = int(sys.argv[1])

if num <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
```
